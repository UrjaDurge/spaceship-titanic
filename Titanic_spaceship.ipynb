# Impot Required Libraries

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder,StandardScaler

from pycaret.classification import setup,compare_models
from sklearn.ensemble import GradientBoostingClassifier
```

# Load Train And Test Data

```python
train=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\UD\\spaceship-titanic\\train.csv")
test=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\UD\\spaceship-titanic\\test.csv")
```

```python
train
```

```python
test
```

# Merge Train And Test Data For Preprocessing

```python
target=train.Transported
test_id=test.PassengerId

train_1=train.drop(['PassengerId','Cabin','Name','Transported'],axis=1)
test_1=test.drop(['PassengerId','Cabin','Name'],axis=1)

data=pd.concat([train_1,test_1],axis=0).reset_index(drop=True)
```

```python
data
```

# Null Value Treatment

```python
data.isna().sum()
```

```python
data.dtypes
```

# Null Value Treatment For Categorical Columns

Filling the numeric values using Mode for Age column and filling 0 in place of null values for other columns

```python
data.Age=data.Age.fillna(data.Age.mean())
```

```python
data[['RoomService','FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = data[['RoomService','FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(value=0)
```

# Null Value Treatment For Categorical Columns

Filling the categorical values using Mode

```python
for column in[
    'HomePlanet',
    'CryoSleep',
    'Destination',
    'VIP'
]:
    data[column]=data[column].fillna(data[column].mode()[0])
```

```python
data.isna().sum()
```

```python
data
```

# Encoding The Categorical Data

Using LabelEncoder for CryoSleep and VIP columns as they are having Ordinal Data

```python
le=LabelEncoder()

data.CryoSleep=le.fit_transform(data.CryoSleep)
data.VIP=le.fit_transform(data.VIP)
```

```python
data
```

Using OneHotEncoder for HomePlanet and Destination columns as they are having nominal data

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
data=data.drop(['HomePlanet_Mars','Destination_55 Cancri e'],axis=1)
```

```python
data
```

# Scaling The Data

```python
scaler=StandardScaler()
scaler.fit(data)

data=pd.DataFrame(scaler.transform(data), index=data.index, columns=data.columns)
data
```

# Splitting The Train And Test Data

```python
train_final=data.loc[:train.index.max(),:].copy()
test_final=data.loc[train.index.max()+1:,:].reset_index(drop=True).copy()
```

```python
train_final
```

```python
test_final
```

```python
target
```

# Model Selection Using Pycaret Library

```python
_=setup(data=pd.concat([train_final,target],axis=1),target='Transported')
```

```python
compare_models()
```

# Model Training

```python
gbc=GradientBoostingClassifier()

gbc.fit(train_final,target)
```

# Making Final Predictions On Test Data

```python
pred=gbc.predict(test_final)
```

```python
submission=pd.concat([test_id,pd.Series(pred,name='Transported')],axis=1)
```
